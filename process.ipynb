```python
import json

data = json.load(open("./data/unique_values.json", "r", encoding="utf-8"))

```

```python
# 将data字典中的元素全部转为字符串
new_data = {str(key): [str(item) for item in value] for key, value in data.items()}
for key, value in new_data.items():
    for item in value:
        try:
            # 去除字符0和以.0结尾的
            if item == "0":
                new_data[key].remove(item)
            elif item.endswith(".0"):
                # 替换为不带.0的
                new_data[key].remove(item)
                new_data[key].append(item[:-2])
        except Exception as e:
            print(e)
            print(f"[ERROR] 处理键 {key} 的值 {item} 时出错: {e}")

with open("./data/映射词表.json", "w", encoding="utf-8") as f:
    json.dump(new_data, f, ensure_ascii=False, indent=4)
```
